```python
import numpy as np
import math

class MinHeap:

    def __init__(self, capacity):
        self.__max_size = capacity 
        self.__arr = np.zeros(capacity + 1, dtype = int)
        self.__heap_indx = 0

    def insert(self, num):
        if self.__heap_indx < self.__max_size:
            self.__heap_indx += 1
            self.__arr[self.__heap_indx] = num
            self.swim(self.__heap_indx)

        else:
            print("No more space")


    def swim(self, indx):
        if indx == 1:
            return
        
        else:
            if self.__arr[indx] < self.__arr[math.floor(indx/2)]:
                self.__arr[indx], self.__arr[math.floor(indx/2)] = self.__arr[math.floor(indx/2)], self.__arr[indx]
                self.swim(math.floor(indx/2))
            
            else:
                return
            
    def extractMin(self):
        min = self.__arr[1]
        self.__arr[1] = self.__arr[self.__heap_indx]
        self.__arr[self.__heap_indx] = 0
        self.__heap_indx -= 1
        self.sink(1)

        return min 
    

    def sink(self, indx):
        left = indx*2
        right = indx*2 + 1
        small = indx

        if left > self.__heap_indx:
            return
        
        else:
            if right <= self.__heap_indx and self.__arr[right] < self.__arr[small]:
                small = right
            
            if left <= self.__heap_indx and self.__arr[left] < self.__arr[small]:
                small = left
            

            if small != indx:
                self.__arr[indx], self.__arr[small] = self.__arr[small], self.__arr[indx]
                self.sink(small)


    def sort(self):
        result = np.zeros(self.__max_size, dtype= int)
        for i in range(self.__max_size):
            num = self.extractMin()
            result[self.__heap_indx] = num
        return result 

    def print_heap(self):
        print(self.__arr) 

    def get_arr(self):
        return self.__arr 
    
    
    def get_heap_index(self):
        return self.__heap_indx 
```

```python
class MaxHeap:

    def __init__(self, capacity):
        self.__max_size = capacity 
        self.__arr = np.zeros(capacity + 1, dtype = int)
        self.__heap_indx = 0

    def insert(self, num):
        if self.__heap_indx < self.__max_size:
            self.__heap_indx += 1
            self.__arr[self.__heap_indx] = num
            self.swim(self.__heap_indx)

        else:
            print("No more space")

    def swim(self, indx):
        if indx == 1:
            return
        
        else:
            if self.__arr[indx] > self.__arr[math.floor(indx/2)]:
                self.__arr[indx], self.__arr[math.floor(indx/2)] = self.__arr[math.floor(indx/2)], self.__arr[indx]
                self.swim(math.floor(indx/2))
            
            else:
                return
            
    def extractMax(self):
        max = self.__arr[1]
        self.__arr[1] = self.__arr[self.__heap_indx]
        self.__arr[self.__heap_indx] = 0
        self.__heap_indx -= 1
        self.sink(1)

        return max 
        

    def sink(self, indx):
        left = indx*2
        right = indx*2 + 1
        big = indx

        if left > self.__heap_indx:
            return
        
        else:
            if left <= self.__heap_indx and self.__arr[left] > self.__arr[big]:
                big = left
            
            if right <= self.__heap_indx and self.__arr[right] > self.__arr[big]:
                big = right

            if big != indx:
                self.__arr[indx], self.__arr[big] = self.__arr[big], self.__arr[indx]
                self.sink(big)


    def sort(self):
        result = np.zeros(self.__max_size, dtype= int)
        for i in range(self.__max_size):
            num = self.extractMax()
            result[self.__heap_indx ] = num
        return result 

    def print_heap(self):
        print(self.__arr)

    def get_heap_index(self):
        return self.__heap_indx 
```

```python
def work_distributin(work, machine):
    heap = MinHeap(machine)
    for i in range(machine):
        heap.insert(0)
    for i in work:
        temp = heap.extractMin()
        new = temp + i
        heap.insert(new)

    return heap.get_arr()[1::] 

work_distributin([2, 4, 7, 1, 6], 4)
```

```python
def th_max(arr, num):
    heap = MaxHeap(len(arr))
    res = np.zeros(num, dtype= int)
    for i in arr:
        heap.insert(i)

    for i in range(num):
        res[i] = heap.extractMax()


    return res

th_max([4,10,2,8,6,7], 3)
```
