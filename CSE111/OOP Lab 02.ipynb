```python
#Task 1
def hospital_fee(**kwargs):
    max_fee = 0
    payer = ""
    for key, value in kwargs.items():
        if value > max_fee:
            max_fee = value
            payer = key
        if (value == max_fee) and str(key) not in payer:
            payer = payer + ", " + key
    return max_fee, payer
max_amount, max_payer  = hospital_fee(Neymar = 1000, Dembele = 600, Reus = 500, Bale = 1000)
print(f"Highest fee was {max_amount} tk which was paid by {max_payer}.")
```

```python
#Task 2
def is_james_bond(lst):
    code, flag = "", False
    for i in range(len(lst)):
        if lst[i] == 0 or lst[i] == 7:
            code += str(lst[i])
    return True if "007" in code else False
is_james_bond([1, 7, 0, 7, 7, 4, 0, 7])
```

```python
#Task 3
def assign_students_to_sections(sections, *students):
    u_dict = {i:[] for i in sections}
    print(sections)
    for name in students:
        value = 0
        for j in range(len(name)):
            value += ord(name[j])
        index = value % len(sections)
        u_dict[sections[index]] += [name]
    return u_dict
print(assign_students_to_sections ('ABCDE', 'Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'))
```

```python
#Task 4
def username_generator(first, last, id, middle=""):
    username = first[:3].upper() + middle + last[len(last)-3:].lower() + "_" + str(id)[len(str(id))-4:]
    return username
first_name, middle_name, last_name, student_id = input("First Name:"), input("Middle Name:"), input("Last Name:"), int(input ("Student ID:"))
print(username_generator(first_name, last_name, student_id, middle_name))
```

```python
#Task 5
def key_generator(*employees):
    key_lst = []
    for name in employees:
        key = name[0].lower()
        reverse = name[-2:0:-1]
        for i in range(len(reverse)):
            key += str(ord(reverse[i]))
        key += name[len(name)-1].upper()
        key_lst += [key]
    return key_lst
key_list = key_generator("Alex", "Bob", "Trudy")
print(f"Encrypted Keys: {key_list}")
```

```python
#Task 6
def playRockPaperScissor(rounds):
    import random
    user, computer = 0, 0
    for i in range(rounds):
        u_pick = input("User: ")
        c_pick = random.choice(["rock", "paper", "scissor"])
        print("Computer:", c_pick)
        if u_pick == "rock" and c_pick == "paper":
            computer += 1
        elif u_pick == "paper" and c_pick == "scissor":
            computer += 1
        elif u_pick == "scissor" and c_pick == "rock":
            computer += 1
        elif u_pick == "paper" and c_pick == "rock":
            user += 1
        elif u_pick == "scissor" and c_pick == "paper":
            user += 1
        elif u_pick == "rock" and c_pick == "scissor":
            user += 1
    print(f"Your Score: {user}\nComputer's Score: {computer}")
    if user > computer:
        print("You have won the game!")
    else:
        print("Computer has won the game!")
playRockPaperScissor(int(input()))
```
