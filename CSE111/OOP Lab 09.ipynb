```python
#Task 1
class NikeBangladesh:
    branches = []
    stock = {'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe': 0}
    sold = 0

    def __init__(self, branch):
        self.branch = branch
        self.stock = {'Air Jordan': 0, 'Cortez': 0, 'Zoom Kobe': 0}
        self.sold = 0
        NikeBangladesh.branches.append(branch)

    def restockProducts(self, delivery):
        self.stock = delivery
        for key, value in self.stock.items():
            NikeBangladesh.stock[key] += value

    def productSold(self, sales):
        for key, value in sales.items():
            NikeBangladesh.stock[key] -= value
            NikeBangladesh.sold += value
            self.stock[key] -= value
            self.sold += value

    def details(self):
        print(f"""Nike {self.branch} outlet:
Products Currently Stocked:
{self.stock}
Sold: {self.sold}""")

    @classmethod
    def status(cls):
        print(f"""Nike Bangladesh Status:
Branches Opened:  {cls.branches}
Currently Stocked
{cls.stock}
Sold: {cls.sold}""")

#Driver Code
print("xxxxxxxxxxxxxx1xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
dhaka = NikeBangladesh("Dhaka Banani")
chittagong = NikeBangladesh("Chittagong GEC")
print("xxxxxxxxxxxxxx2xxxxxxxxxxxxxxxx")
dhaka.details()
print("xxxxxxxxxxxxxx3xxxxxxxxxxxxxxxx")
chittagong.details()
print("xxxxxxxxxxxxxx4xxxxxxxxxxxxxxxx")
dhaka.restockProducts(
{"Air Jordan":1200,"Cortez":200,"Zoom Kobe":200})
chittagong.restockProducts(
{"Air Jordan":1000,"Cortez":250,"Zoom Kobe":100})
print("xxxxxxxxxxxxxx5xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
print("xxxxxxxxxxxxxx6xxxxxxxxxxxxxxxx")
dhaka.productSold({"Air Jordan":760,"Cortez":90})
chittagong.productSold({"Air Jordan":520,"Zoom Kobe":70})
print("xxxxxxxxxxxxxx7xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
```

```python
#Task 2
class Player:
    total = 0
    enlisted = []

    def __init__(self, name = None, number = 10, team = None):
        self.name = name
        self.number = number
        self.team = team
        Player.total += 1
        if name and name not in Player.enlisted:
            Player.enlisted.append(self.name)

    def set_name(self, name):
        self.name = name
        if self.name not in Player.enlisted:
            Player.enlisted.append(self.name)

    def set_team(self, team):
        self.team = team

    def set_number(self, number):
        self.number = number

    def player_detail(self):
        return f"""Player Name: {self.name}
Jersey Number: {self.number}
Country: {self.team}"""

    @classmethod
    def info(cls):
        print(f"""Total number of players: {cls.total}
Players enlisted so far: {', '.join(cls.enlisted)}""")

#Driver Code
print("Total number of players:", Player.total)
print("---------------------------")
p1 = Player()
p1.set_name("Neymar")
p1.set_team("Brazil")
print(p1.player_detail())
print('========================')
Player.info()
print("---------------------------")
p2 = Player("Ronaldo")
p2.set_number(7)
p2.set_team("Portugal")
print(p2.player_detail())
print('========================')
Player.info()
print("---------------------------")
p3 = Player("Messi")
p3.set_team("Argentina")
print(p3.player_detail())
print('========================')
Player.info()
print("---------------------------")
p4 = Player("Mbappe", 10, "France")
print(p4.player_detail())
print('========================')
Player.info()
```

```python
#Task 3
class SultansDine:
    branches = []
    t_sales = 0

    def __init__(self, name):
        self.name = name
        self.sales = 0
        SultansDine.branches.append(self)

    def sellQuantity(self, amount):
        if amount < 10:
            self.sales += amount * 300
        elif amount < 20:
            self.sales += amount * 350
        else:
            self.sales += amount * 400
        SultansDine.t_sales += self.sales

    def branchInformation(self):
        print(f"""Branch Name: {self.name}
Branch Sell: {self.sales} Taka""")

    @classmethod
    def details(cls):
        print(f"""Total Number of branch(s): {len(SultansDine.branches)}
Total Sell: {SultansDine.t_sales} Taka""")
        if len(SultansDine.branches) != 0:
            for ref in SultansDine.branches:
                print(f"""Branch Name: {ref.name}, Branch Sell: {ref.sales} Taka
Branch consists of total sell's: {round(ref.sales / SultansDine.t_sales * 100, 2)}%""")

#Driver Code
SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
