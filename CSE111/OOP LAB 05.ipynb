```python
#Task 2
class Farmer:
    def __init__(self, info=None):
        if type(info) == str:
            print(f"Welcome to your farm, {info}!")
        elif type(info) == int:
            print(f"Welcome to your farm. Your farm ID is {info}!")
        else:
            print("Welcome to your farm!")
        self.crops = []
        self.fishes = []
    def addCrops(self, *crops):
        if crops:
            for crop in crops:
                self.crops += [crop]
            print(f"{len(crops)} crop(s) added.")
        else: print("No crop(s) added.")
    def addFishes(self, *fishes):
        if fishes:
            for fish in fishes:
                self.fishes += [fish]
            print(f"{len(fishes)} fish(s) added.")
        else: print("No fish added.")
    def showGoods(self):
        if len(self.crops) != 0:
            print(f"You have {len(self.crops)} crop(s):\n{','.join(self.crops)}")
        else: print("You don't have any crop(s).")
        if len(self.fishes) != 0:
            print(f"You have {len(self.fishes)} crop(s):\n{','.join(self.fishes)}")
        else: print("You don't have any fish(s).")
#Driver Code
f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```

```python
#Task 3
class TaxiLagbe:
    def __init__(self, num, area):
        self.num = num
        self.area = area
        self.passengers = 0
        self.fare = 0
        self.persons = []
    def addPassenger(self, *passengers):
        if (self.passengers + len(passengers)) <= 4:
            self.passengers += len(passengers)
            for info in passengers:
                person = info[:-4]
                self.persons += [person]
                self.fare += int(info[-3:])
                print(f"Dear {person}! Welcome to TaxiLagbe.")
        else: print(f"Taxi Full! No more passengers can be added.")
    def printDetails(self):
        print(f"""Trip info for Taxi number: {self.num}
This taxi can only cover the {self.area} area.
Total passengers: {self.passengers}
Passenger lists:
{", ".join(self.persons)}
Total collected fare: {self.fare} taka""")
#Driver Code
taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200','Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115', 'Parker_215')
print('-------------------------------')
taxi2.printDetails()
```

```python
#Task 4
import math
class Sphere:
    def __init__(self, id, radius=1, color="White"):
        self.id = id
        self.color = color
        self.volume = round(4/3 * math.pi * math.pow(radius, 3), 4)
        if type(radius) == int:
            self.radius = radius
    def printDetails(self):
        print(f"Sphere ID: {self.id}\nColor: {self.color}\nVolume: {self.volume}")
    def merge_sphere(self, *sphere):
        print("Spheres are being merged")
        if len(sphere) == 1:
            self.volume += sphere[0].volume
            if self.color != sphere[0].color:
                self.color = "Mixed Color"
        elif len(sphere) == 2:
            self.volume += sphere[0].volume + sphere[1].volume
            if self.color != (sphere[0].color or sphere[1].color):
                self.color = "Mixed Color"
#Driver Code
sphere1 = Sphere("Sphere 1")
print("1***************")
sphere1.printDetails()
print("2***************")
sphere2 = Sphere("Sphere 2", 3)
print("3***************")
sphere2.printDetails()
print("4***************")
sphere3 = Sphere("Sphere 3", 2)
print("5***************")
sphere3.printDetails()
print("6***************")
sphere3.merge_sphere(sphere1,sphere2)
print("7***************")
sphere3.printDetails()
print("8***************")
sphere4 = Sphere("Sphere 4", 5, "Purple")
print("9***************")
sphere4.merge_sphere(sphere3)
print("10***************")
sphere4.printDetails()
```

```python
#Task 3
class Department:
    def __init__(self, sub="ChE Department", sec=5):
        self.sub = sub
        self.sec = sec
        self.avg = 0
        print(f"The {self.sub} has {self.sec} sections.")

    def add_students(self, *students):
        if len(students) == self.sec:
            for count in students:
                self.avg += count
            self.avg = round(self.avg / len(students), 2)
            print(f"The {self.sub} has an average of {self.avg} students in each section.")
        else: print(f"The {self.sub} doesn't have {len(students)} sections.")

    def merge_Department(self, *location):
        self.avg *= self.sec
        for s in location:
            self.avg += s.avg * s.sec
            print(f"{s.sub} is merged to {self.sub}.")
        self.avg = round(self.avg / self.sec, 2)
        return f"Now the {self.sub} has an average of {self.avg} students in each section."

#Driver Code
d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21,30,40,36,10,32,27,51,45,15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
#Task 4
class Shopidify:
    def __init__(self, name="Guest"):
        print(f"Welcome {name} to Shopidify" if name != "Guest" else "Welcome to Shopidify")
        self.name = name
        self.cart = {}
        self.transaction = 0

    def add_to_cart(self, *item):
        if len(item) == 1:
            if type(item[0]) == list:
                for i in range(0, len(item[0]), 2):
                    if item[0][i] in self.cart.keys():
                        self.cart[item[0][i]] += item[0][i+1]
                    else: self.cart[item[0][i]] = item[0][i+1]
            elif type(item[0]) != list:
                if item[0] in self.cart.keys():
                    self.cart[item[0]] += 1
                else: self.cart[item[0]] = 1
        elif len(item) == 2:
            if item[0] in self.cart.keys():
                self.cart[item[0]] += item[1]
            else: self.cart[item[0]] = item[1]

    def display_cart(self):
        print(f"Items in the cart for {self.name}:")
        for key, value in self.cart.items():
            print(f"- {key} : {value}x")

    def checkout(self):
        print(f"Checkout completed for {self.name}")
        self.transaction += 1

    def display_history(self):
        print(f"Purchase history for {self.name}:\nTransaction {self.transaction}:")
        for key, value in self.cart.items():
            print(f"- {key} : {value}x")

#Driver Code
guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart(["Chocolate Chip Cookies", 3,"Goku Action Figure",2,"Dumbbells-5kg",2])
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
```

```python
#Task 5
class Author:
    def __init__(self, name=None):
        self.name = name
        self.library = {}
        self.books = 0

    def addBook(self, book, genre):
        if self.name != None:
            self.books += 1
            if genre in self.library:
                self.library[genre].append(book)
            else:
                self.library[genre] = [book]
        else: print("A book can not be added without author name")

    def setName(self, name):
        self.name = name

    def printDetail(self):
        print(f"Number of Book(s): {self.books} ")
        print(f"Author Name: {self.name}")
        for key, value in self.library.items():
            print(f"{key}: {', '.join(value)}")

#Driver Code
a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```
