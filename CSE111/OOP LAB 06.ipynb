```python
#Task 1
class Teacher:
    def __init__(self, name, department):
        self.name = name
        self.department = department
        self.courses = []

    def addCourse(self, location):
        self.courses.append(location.course)

    def printDetail(self):
        print("====================================")
        print(f"Name: {self.name}\nDepartment: {self.department}\nList of courses")
        print("====================================")
        print("\n".join(self.courses))
        print("====================================")
class Course:
    def __init__(self, course):
        self.course = course

#Driver Code
t1 = Teacher("Saad Abdullah", "CSE")
t2 = Teacher("Mumit Khan", "CSE")
t3 = Teacher("Sadia Kazi", "CSE")
c1 = Course("CSE 110 Programming Language I")
c2 = Course("CSE 111 Programming Language-II")
c3 = Course("CSE 220 Data Structures")
c4 = Course("CSE 221 Algorithms")
c5 = Course("CSE 230 Discrete Mathematics")
c6 = Course("CSE 310 Object Oriented Programming")
c7 = Course("CSE 320 Data Communications")
c8 = Course("CSE 340 Computer Architecture")
t1.addCourse(c1)
t1.addCourse(c2)
t2.addCourse(c3)
t2.addCourse(c4)
t2.addCourse(c5)
t3.addCourse(c6)
t3.addCourse(c7)
t3.addCourse(c8)
t1.printDetail()
t2.printDetail()
t3.printDetail()
```

```python
class Student:
    def __init__(self, name, id, cgpa):
        self.name = name
        self.id = id
        self.cgpa = cgpa

    def setId(self, id):
        self.id = id

class Department:
    def __init__(self, department):
        self.department = department
        self.info = {}

    def findStudent(self, id):
        if id in self.info:
            print(f"Student info:\nStudent Name: {self.info[id][0]}\nID: {id}\nCGPA:  {self.info[id][1]}""")
        else: print("Student with this ID doesn't exist, Please give a valid ID")

    def addStudent(self, *args):
        for obj in args:
            if obj.id not in self.info:
                self.info[obj.id] = [obj.name, obj.cgpa]
                print(f"Welcome to CSE department, {obj.name}")
            else: print("Student with the same ID already exists, Please try with another ID")

    def details(self):
        print(f"Department Name: {self.department}\nNumber of student: {len(self.info.keys())}\nDetails of the students:")
        for key, value in self.info.items():
            print(f"Student name: {value[0]}, ID: {key}, cgpa: {value[1]}")

#Driver Code
s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1,s2,s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib",22301010,3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib",22201010,2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
```

```python
#Task 3
class Vaccine:
    def __init__(self, vaccine, country, date):
        self.vaccine = vaccine
        self.country = country
        self.date = date

class Person:
    def __init__(self, name, age, occupation="General Citizen"):
        self.name = name
        self.age = age
        self.occupation = occupation
        self.dose = None
        self.dose2 = None

    def pushVaccine(self, obj, dose=None):
        if self.age >= 25 or self.occupation == "Student":
            if not dose:
                self.dose = obj.vaccine
                self.dose2 = obj.date
                print(f"1st dose done for {self.name}")
            elif self.dose == obj.vaccine:
                self.dose2 = None
                print(f"2nd dose done for {self.name}")
            else:
                print(f"Sorry {self.name}, you canâ€™t take 2 different vaccines")
        else:
            print(f"Sorry {self.name}, Minimum age for taking vaccines is 25 years now.")

    def showDetail(self):
        print(f"Name: {self.name} Age: {self.age} Type: {self.occupation}")
        print(f"Vaccine name: {self.dose}")
        if self.dose and self.dose2:
            print(f"1st dose: Given\n2nd dose: Please come after {self.dose2} days")
        elif self.dose and not self.dose2:
            print(f"1st dose: Given\n2nd dose: Given")

#Driver Code
astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
