```python
#Task 11
class Transport:
    total_traveller = 0

    def __init__(self, name, fare):
        self.name = name
        self.baseFare = fare

    def __str__(self):
        s =  "Name: "+self.name+", Base fare: "+str(self.baseFare)
        return s
class Bus(Transport):
    def  __init__(self, name, fare):
        super().__init__(name,fare)
        self.passenger = {}
        print(f"Base-fare of {self.name} is {self.baseFare} Taka")

    def addPassengerWithBags(self, *args):
        for j in range(0, len(args) - 1, 2):
            if args[j+1] > 5:
                self.passenger[args[j]] = self.baseFare + 105
            elif 3 <=  args[j+1] <= 5:
                self.passenger[args[j]] = self.baseFare + 60
            else:
                self.passenger[args[j]] = self.baseFare
        Transport.total_traveller += len(self.passenger.keys())
    def __str__(self):
        temp = f"{super().__str__()}\nTotal Passenger(s): {len(self.passenger.keys())}\nPassenger details:\n"
        for key, value in self.passenger.items():
            temp += f"Name: {key}, Fare: {value}\n"
        return temp
class Train(Transport):
    def  __init__(self, name, fare):
        super().__init__(name,fare)
        self.passenger = {}
        print(f"Base-fare of {self.name} is {self.baseFare} Taka")

    def addPassengerWithBags(self, *args):
        for j in range(0, len(args) - 1, 2):
            if args[j+1] > 5:
                self.passenger[args[j]] = self.baseFare + 105
            elif 3 <=  args[j+1] <= 5:
                self.passenger[args[j]] = self.baseFare + 60
            else:
                self.passenger[args[j]] = self.baseFare
        Transport.total_traveller += len(self.passenger.keys())
    def __str__(self):
        temp = f"{super().__str__()}\nTotal Passenger(s): {len(self.passenger.keys())}\nPassenger details:\n"
        for key, value in self.passenger.items():
            temp += f"Name: {key}, Fare: {value}\n"
        return temp

t1 = Bus("Volvo", 950)
print("=================================")
t1.addPassengerWithBags("David", 6,  "Mike", 1, "Carol", 3)
print(t1)
print("=================================")
t2 = Train("Silk City", 850)
print("=================================")
t2.addPassengerWithBags("Bob", 2, "Simon", 4)
print("=================================")
print(t2)
print("=================================")
print("Total Passengers in Transport: ", Transport.total_traveller )
```

```python
class User:
    def __init__(self, name, destination, ride="Single"):
        self.name = name
        self.destination = destination
        self.ride = ride
        self.stat = f"Status: {self.name} is looking for a {self.ride.lower()} ride!"

    def status(self):
        print(self.stat)

class Uber:
    def __init__(self, car_number, car_type, *routes):
        self.car_number = car_number
        self.car_type = car_type
        self.routes = routes
        self.passengers = []

    def details(self):
        print(f"Car number: {self.car_number}")
        print(f"Type: {self.car_type}")
        print("Routes:", " --> ".join(self.routes))

    def pick(self, *users):
        for user in users:
            if self.car_type == "Shared" and user.ride == "Shared":
                if user.destination in self.routes:
                    print(f"{user.name} has been picked up.")
                    self.passengers.append(user)
                    self.status = f"Status: {user.name} boarded in car {self.car_number}!"
                else:
                    print(f"{user.name}'s destination is different from this car's route.")
            elif user.ride == "Single":
                print(f"{user.name} is looking for a different ride.")

user1 = User("Brooks", "Banani", "Shared")
user2 = User("Jocelyn", "Uttara")
user3 = User("Robert", "Gulshan", "Shared")
user4 = User("Langdon", "Mohakhali", "Shared")


user1.status()
user2.status()
user3.status()
user4.status()
print("----------------------------------")
car1 = Uber("0K32BH", "Shared", "Mohakhali", "Banani", "Nikunja", "Uttara")
car1.details()
print("----------------------------------")
car1.pick(user1,user2,user3,user4)
print("----------------------------------")
user1.status()
user2.status()
user3.status()
user4.status()
print("----------------------------------")
car2 = Uber("5GD2BD", "Single", "Uttara")
car3 = Uber("4T12FR", "Shared", "Gulshan", "Bashundhara")
car2.details()
car3.details()
print("----------------------------------")
car2.pick(user2, user3)
print("----------------------------------")
car3.pick(user3)
print("----------------------------------")
user2.status()
user3.status()
```
