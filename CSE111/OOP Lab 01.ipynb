```python
#Task 1
user_list = []
while True:
    user_input = input()
    if user_input != "STOP":
         user_list.append(user_input)
    else:
         break
counter = 1
for i in range(len(user_list)):
    for j in range(i+1, len(user_list)):
        if user_list[i] == user_list[j]:
            counter += 1
            user_list[j] = "X"
    if user_list[i] != "X":
        print(f"{user_list[i]} - {counter} times")
    counter = 1
```

```python
#Task 2
import math
total_lists = int(input())
sum = 0
index = 0
max = -math.inf
final_list = []
sum_list = []
for count in range(total_lists):
    user_list = input().split(" ")
    final_list.append([int(num) for num in user_list])
for i in range(len(final_list)):
    for j in range(len(final_list[i])):
        sum += final_list[i][j]
    sum_list.append(sum)
    if sum > max:
        max = sum
        index = i
    sum = 0
print(sum_list[index], final_list[index], sep="\n")
```

###Dictionary & Tuple

```python
#Task 1
def convert_to_dict(var):
    var = var.split(", ")
    u_dict = {}
    for i in var:
        j,k = i.split(": ")
        u_dict[j] = int(k)
    return u_dict


def num_sort(var):
    for i in range(len(var)):
        min_i = i
        for j in range(i+1, len(var)):
            if var[min_i] > var[j]:
                min_i = j
        var[i], var[min_i] = var[min_i], var[i]
    return var


dict_one = convert_to_dict(input())
dict_two = convert_to_dict(input())
final = {}
temp = []
for i, j in dict_one.items():
    if i not in final:
        final[i] = j
    else:
        final[i] += j
for i, j in dict_two.items():
    if i not in final:
        final[i] = j
    else:
        final[i] += j
for k in final.values():
    if k not in temp:
        temp += [k]
print(f"""{final}
Values: {tuple(num_sort(temp))}""")
```

```python
#Task 3
def conv_dict(var):
    var = var.split(", ")
    u_dict = {}
    for i in var:
        j,k = i.split(" : ")
        u_dict[j] = k
    return u_dict

user_dict = conv_dict(input())
inverse = {}
for key,value in user_dict.items():
    if value not in inverse:
        inverse[value] = [key]
    else:
        inverse[value] += [key]
print(inverse)
```

```python
#Task 6
keypad = {1: '.,?!:', 2: 'ABC', 3: 'DEF', 4: 'GHI', 5: 'JKL', 6: 'MNO', 7: 'PQRS', 8: 'TUV', 9: 'WXYZ', 0: ' '}
user_input = input()
modified = ""
cipher = ""
for char in user_input:
    if "a" <= char <= "z":
        modified += chr(ord(char)-32)
    else:
        modified += char
for i in modified:
    for key, character in keypad.items():
        if i in character:
            clicks = 1
            for j in character:
                if i == j:
                    break
                else:
                    clicks += 1
            cipher += str(key) * clicks
print(cipher)
```
